## Training simulation datasets sampled within a unit square

```python
fig, axs = plt.subplots(1,5,figsize=(5*5,5))

for i in range(5):
    axs[i].scatter(ib.train_X[i][:,0],ib.train_X[i][:,1], c=ib.train_y[i], cmap=cmap)
    axs[i].set_title(ib.dtype[i])
    axs[i].add_patch(Rectangle(xy=(-1,-1), width=2, height=2, edgecolor='k', fill=False))
    axs[i].set_xlim([-3,3]);
    axs[i].set_ylim([-3,3]);
```

## Training simulation datasets sampled within a unit circle

```python
fig, axs = plt.subplots(1,5,figsize=(5*5,5))

for i in range(5):
    axs[i].scatter(ib.Ctrain_X[i][:,0],ib.Ctrain_X[i][:,1], c=ib.Ctrain_y[i], cmap=cmap)
    axs[i].set_title(ib.dtype[i])
    axs[i].add_patch(Circle((0,0), radius=1, edgecolor='k', fill=False))
    axs[i].set_xlim([-3,3]);
    axs[i].set_ylim([-3,3]);
```
